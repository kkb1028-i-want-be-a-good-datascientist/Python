### 넘파이 사용

```python
#sin-1.py

import math
import matplotlib.pyplot as plt

T = range(100)
X = [(2*math.pi*t)/len(T) for t in T]
Y = [math.sin(value) for value in X]

plt.plot(X,Y)
plt.show()
```

```python
#sin-2.py

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 100)
Y = np.sin(X)

plt.plot(X,Y)
plt.show()
```

### 넘파이 패키지

```python
# 이항식 : x^2 - 2x + 1

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-3, 2, 200)
Y = X**2 - 2*X + 1

plt.plot(X,Y)
plt.show()
```

### 다중 곡선 그리기

```python
# [0, 2*pi] 간격의 sin(x)과 cos(x)을 둘 다 보여주기

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X,Ya)
plt.plot(X,Yb)
plt.show()
```

```python
# 지연 랜더링, deferred rendering 메커니즘은 맷플롯립의 핵심

import numpy as np
import matplotlib.pyplot as plt

def plot_slope(X,Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys/Xs)
    
X = np.linspace(-3, 3, 100)
Y = np.exp(-X**2)

plt.plot(X,Y)
plot_slope(X,Y)
plt.show()
```

```python
import matplotlib.pyplot as plt

X, Y = [],[]
for line in open('my_data.txt', 'r'):
    values = [float(s) for s in line.split()]
    X.append(values[0])
    Y.append(values[1])
    
plt.plot(X,Y)
plt.show()
```

```python
import matplotlib.pyplot as plt

with open('my_data.txt', 'r') as f:
    X, Y = zip(*[[float(s) for s in line.split()] for line in f])
    
plt.plot(X,Y)
plt.show()
```

### 다중 막대 차트 그리기

```python
# 여러 수량을 비교
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

X = np.arange(4)
plt.bar(X + 0.00, data[0], color='b', width=0.25)
plt.bar(X + 0.25, data[1], color='g', width=0.25)
plt.bar(X + 0.50, data[2], color='r', width=0.25)

plt.show()
```

```python
# 반복자 enumerate 는 현재 행과 해당 첨자를 둘 다 반환
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

color_list = ['b', 'g', 'r']
gap = .8 / len(data)
for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i*gap, row, 
           width = gap, 
           color = color_list[i % len(color_list)])

plt.show()
```

```python
import matplotlib.pyplot as plt

A = [5., 30., 45., 22.]
B = [5., 25., 50., 20.]

X=range(4)

plt.bar(X, A, color='b')
plt.bar(X, B, color='r', bottom=A)
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
import numpy as np
import matplotlib.pyplot as plt

A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1.,  2.,  1.,  1.])
X = range(4)

plt.bar(X,A,color='b')
plt.bar(X,B,color='g', bottom=A)
plt.bar(X,C,color='r', bottom=A+B)
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
import numpy as np
import matplotlib.pyplot as plt

data = np.array([[5., 30., 45., 22.],
                 [5., 25., 50., 20.],
                 [1.,  2.,  1.,  1.]        
    ])
color_list = ['b','g','r']

X=np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(X, data[i],
           bottom = np.sum(data[:i], axis=0),
           color=color_list[i % len(color_list)])
plt.show()
```

### 양방향 막대 차트 그리기

```python
# 두 막대 차트를 동시에 양방향으로 표시
import numpy as np
import matplotlib.pyplot as plt

women_pop = np.array([5., 30., 45., 22.])
men_pop   = np.array([5., 25., 50., 20.])
X = np.arange(1,5)

plt.barh(X, women_pop, color='r')
plt.barh(X, -men_pop, color='b')
plt.show()
```

### 삼각화 그리기

```python
# 공간 위치를 처리할 때 삼각화 생긴다. 점과 이웃 관계 사이의 거리를 보여줌
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri

data = np.random.rand(100,2)

triangles = tri.Triangulation(data[:,0], data[:,1])

plt.triplot(triangles)
plt.show()
```
